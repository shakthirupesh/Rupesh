### Q6. Find the three quartiles and the interquartile range (IQR).

```python
# Sort the CEO ages data
sorted_ages = sorted(ages)

# Calculate the median (Q2)
n = len(sorted_ages)
Q2 = (sorted_ages[(n - 1) // 2] + sorted_ages[n // 2]) / 2

# Calculate the first quartile (Q1)
lower_half = sorted_ages[:n // 2]
n1 = len(lower_half)
Q1 = (lower_half[(n1 - 1) // 2] + lower_half[n1 // 2]) / 2

# Calculate the third quartile (Q3)
upper_half = sorted_ages[n // 2:]
n2 = len(upper_half)
Q3 = (upper_half[(n2 - 1) // 2] + upper_half[n2 // 2]) / 2

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

print("Q1 (First Quartile):", Q1)
print("Q2 (Median - Second Quartile):", Q2)
print("Q3 (Third Quartile):", Q3)
print("IQR (Interquartile Range):", IQR)
```

### Q17. A data scientist wants to estimate with 95% confidence the proportion of CEOs of Fortune 500 companies are above 60 years in the population.
Another recent study showed that 25% of CEOs interviewed were above 60. The data scientist wants to be accurate within 2% of the true proportion. Find the minimum sample size necessary.

```python
confidence_level = 0.95
margin_of_error = 0.02
proportion = 0.25 

# Calculate the critical value for a 95% confidence level
critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

sample_size = (critical_value**2 * proportion * (1 - proportion)) / margin_of_error**2

sample_size = int(sample_size + 0.5)

print("Minimum Sample Size:", sample_size)
```

### Q18. The same data scientist wants to estimate the true proportion of CEOs ascending to the post and above 60 years. She wants to be 90% confident and accurate within 5% of true proportion. Find the minimum sample size necessary.

```python
confidence_level = 0.90
margin_of_error = 0.05

z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the minimum sample size using the formula for proportion estimation
sample_size = (z_score**2 * 0.25) / margin_of_error**2

sample_size = int(sample_size + 0.5)

print("Minimum Sample Size:", sample_size)
```

### 18. With the help of hypothesis testing check whether fraudulent transactions of higher value than normal transactions?

Null Hypothesis (H0): There is no significant difference in the transaction amounts between fraudulent and normal transactions.

H0: μ_fraud = μ_normal
Alternative Hypothesis (H1): Fraudulent transactions have a higher mean transaction amount compared to normal transactions.

H1: μ_fraud > μ_normal

```python
from scipy.stats import ttest_ind
alpha = 0.05 
fraudulent = cd[cd['Class'] == 1]['Amount']
normal = cd[cd['Class'] == 0]['Amount']
```

```python
ttest_ind(fraudulent, normal, alternative='greater', equal_var=True)
```

- t_statistic=3.0055523
- pvalue=0.0013256
- significance level= 0.05
- pvalue < significance level
- Reject the null hypothesis: Fraudulent transactions have a higher mean amount than normal transactions.

#### What columns seem to have outliers based on min, max and percentile values, IQR range along with the standard deviation and mean absolute deviation?

```python
cd = pd.read_csv('creditcard.csv')

# Set the threshold for each method
z_score_threshold = 3
iqr_threshold = 1.5
mad_threshold = 3

# Initialize lists to store columns with potential outliers
outliers_min_max = []
outliers_percentile = []
outliers_iqr = []
outliers_z_score = []
outliers_mad = []

# Loop through columns to identify outliers
for column in cd.columns:
    if cd[column].dtype in ['float64', 'int64']:
        # Min-Max method
        min_val = cd[column].min()
        max_val = cd[column].max()
        min_max_outliers = cd[(cd[column] < min_val) | (cd[column] > max_val)]
        if not min_max_outliers.empty:
            outliers_min_max.append(column)
        
        # Percentile method (0.01% and 99.99% as an example)
        p01 = np.percentile(cd[column], 0.01)
        p9999 = np.percentile(cd[column], 99.99)
        percentile_outliers = cd[(cd[column] < p01) | (cd[column] > p9999)]
        if not percentile_outliers.empty:
            outliers_percentile.append(column)

        # IQR method
        Q1 = cd[column].quantile(0.25)
        Q3 = cd[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - iqr_threshold * IQR
        upper_bound = Q3 + iqr_threshold * IQR
        iqr_outliers = cd[(cd[column] < lower_bound) | (cd[column] > upper_bound)]
        if not iqr_outliers.empty:
            outliers_iqr.append(column)
        
        # Z-Score method
        z_scores = (cd[column] - cd[column].mean()) / cd[column].std()
        z_outliers = cd[abs(z_scores) > z_score_threshold]
        if not z_outliers.empty:
            outliers_z_score.append(column)
        
        # Mean Absolute Deviation (MAD) method
        mad = abs(cd[column] - cd[column].median())
        mad_value = 1.4826 * np.median(mad)
        mad_outliers = cd[abs(mad) > mad_threshold * mad_value]
        if not mad_outliers.empty:
            outliers_mad.append(column)

# Print columns with potential outliers using different methods
print("Columns with potential outliers (Min-Max method):")
print(outliers_min_max)
print("Columns with potential outliers (Percentile method):")
print(outliers_percentile)
print("Columns with potential outliers (IQR method):")
print(outliers_iqr)
print("Columns with potential outliers (Z-Score method):")
print(outliers_z_score)
print("Columns with potential outliers (MAD method):")
print(outliers_mad)
```
